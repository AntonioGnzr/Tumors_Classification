```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#Opening the Dataset
df = pd.read_csv(r"Dataset_Tumors.csv")
df
```

```python
#Renaming the age variable
df['age']=df['age'].replace("'>=60'",'old')
df['age']=df['age'].replace("'30-59'",'middle')
df['age']=df['age'].replace("'<30'",'young')
df['class']=df['class'].str.replace('\'','')
df
```

```python
#Exploring the age class
df['age'].value_counts()
```

```python
#Encoding missing values in the standard format using NaN
df=df.replace('?',np.NaN)
df
```

```python
#For each variable, counting the number of missing variables
df.isnull().sum()
```

```python
#Decided to drop some variables,reducing the number of variables instead of the number of observations that is still low
df=df.drop('degree-of-diffe',axis=1)
df=df.drop('histologic-type',axis=1)
df=df.dropna(axis=0)
df
```

```python
#Exploring the different classes
df['class'].value_counts()
#Some class have really low number of observations, it may be difficult to calssificate them
```

```python
#I have to drop classes with only one observation because it wouldnt have any explanatory power
to_drop=df['class'].value_counts()[df['class'].value_counts()<15].index
df=df[~df['class'].isin(to_drop)]
df
```

```python
#Encoding Variables so that they are ready to be trained
#Started by encoding age with an ordinal encoding, as the order and the distance is important
from sklearn.preprocessing import OrdinalEncoder
categories =[['young','middle','old']]
enc_ord = OrdinalEncoder()
enc_ord.fit(df.loc[:,['age']])
df.loc[:,['age']]=enc_ord.transform(df.loc[:,['age']])
df
```

```python
#The other variables are encoded as dummy variables
from sklearn.preprocessing import OneHotEncoder
f = df.columns.get_loc
col_ohe = df.iloc[:,np.r_[f('sex'), f('bone'):f('class')]]
enc_ohe = OneHotEncoder(drop='first')
enc_ohe.fit(col_ohe)
enc_ohe.transform(col_ohe).toarray()
df_enc=pd.DataFrame(data=enc_ohe.transform(col_ohe).toarray(),columns=enc_ohe.get_feature_names(col_ohe.columns))
df_enc
```

```python
#I need to reset the index to merge the different econded variables
df=df.reset_index()
df
```

```python
#Merging columns
df_enc['age']=df['age']
df_enc['class']=df['class']
df_enc
```

```python
#Separating explanatory variable and target variable
X = df_enc.iloc[:,:-1]
y = df_enc.iloc[:,-1]
```

```python
#Separating training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=13, stratify = y)
```

```python
#Trying k-neighbord classification
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier()
parameters = {'n_neighbors':np.arange(1,11)}
```

```python
from sklearn.model_selection import GridSearchCV
neigh=GridSearchCV(neigh, parameters,cv=5,scoring='accuracy', verbose=50, n_jobs=-1)
```

```python
neigh.fit(X_train,y_train)
```

```python
y_pred=neigh.best_estimator_.predict(X_test)
y_pred
```

```python
#Relevant metrics
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#Confusion matrix
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
plot_confusion_matrix(neigh, X_test, y_test)
plt.show()
```

```python
#Random forest from scikit learn
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
parameters = {
    'criterion':['gini'],
    'max_depth':[4],
    'min_samples_split':[5,10],
    'min_samples_leaf':[2],
    'class_weight':['balanced']}

rfc=GridSearchCV(rfc, parameters,cv=5,scoring='accuracy', verbose=50, n_jobs=-1)
rfc.fit(X_train, y_train)
```

```python
y_pred_forest = rfc.predict(X_test)
y_pred_forest
```

```python
#Relevant metrics
print(classification_report(y_test,y_pred_forest))
```

```python
#Confusion matrix
plot_confusion_matrix(rfc, X_test, y_test)
plt.show()
```

```python
#Naive Bayesian Classification
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
parameters = {'alpha':[0.01,0.1,1,2,3,4,5]}

mnb=GridSearchCV(mnb, parameters,cv=5,scoring='accuracy', verbose=50, n_jobs=-1)
mnb.fit(X_test, y_test)
```

```python
y_pred_mnb=mnb.predict(X_test)
y_pred_mnb
```

```python
#Relevant metrics
print(classification_report(y_test,y_pred_mnb))
```

```python
#Confusion matrix
plot_confusion_matrix(mnb, X_test, y_test)
plt.show()
```

The last model is the best model. Particularly it is able to classify breast, gallbladder, head and neck, ovary and lung cancer with a high f1 score.
However the dataset suffers from a low number of observations and unbalanced classes, thus the even if it has been used a cross-validation during the trainig, the model may vary a lot based on the randon_state used.
